```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import warnings
warnings.filterwarnings('ignore')

data = load_boston()
```

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
dt = DecisionTreeRegressor()
```

```python
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
dt.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(mean_absolute_error(y_pred_lr, y_test))
print(mean_absolute_error(y_pred_rf, y_test))
print(mean_absolute_error(y_pred_dt, y_test))
```

```python
print(r2_score(y_pred_lr, y_test))
print(r2_score(y_pred_rf, y_test))
print(r2_score(y_pred_dt, y_test))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred_dt, y_test)
plt.plot(dt.predict(X_train), y_train, c='r')
plt.title("Decision_Tree")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred_rf, y_test)
plt.plot(dt.predict(X_train), y_train, c='g')
plt.title("Random_Forest")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred_lr, y_test)
plt.plot(dt.predict(X_train), y_train, c='k')
plt.title("Linear_regression")
plt.show()
```
